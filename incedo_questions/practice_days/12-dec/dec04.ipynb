Question 6 - Write a function that takes a list of numbers and returns the maximum value.<br>
Example - find_max([3, 1, 4, 1, 5]) -> 5

```python
input_nums = [3, 1, 4, 1, 5]
def find_max(input_list: list[int]) -> int:
    input_list.sort()
    return input_list[-1]


print(f"The maximum from the list of number {input_nums} is {find_max(input_nums)}")
```

Q5. Write a function that counts the number of vowels (a, e, i, o, u) in a string. <br>
Example - count_vowels("hello") -> 2

```python
vowel_list = ['a', 'e', 'i', 'o', 'u']

def count_vowels(input_word: str) -> int:
    vowel_count = 0
    for char in input_word:
        if char in vowel_list:
            vowel_count += 1
    return vowel_count
input_str = "helloooo"
print(f"The vowel count of {input_str} is {count_vowels(input_str)}")
```
