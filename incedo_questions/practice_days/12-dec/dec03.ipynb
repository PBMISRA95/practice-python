### Question 8 - Write a function that removes duplicates from a list and returns the result.
Example - remove_duplicates([1, 2, 2, 3, 4, 4]) -> [1, 2, 3, 4]

```python
input_list = [1, 2, 2, 3, 4, 4]

def remove_duplicates(input_list: list[int]) -> list[int]:
    output_list = []
    for num in input_list:
        if num in output_list:
            continue
        else:
            output_list.append(num) 
    return output_list

print(f"\n The given list {input_list} after removing duplicates is {remove_duplicates(input_list)} ")
```
