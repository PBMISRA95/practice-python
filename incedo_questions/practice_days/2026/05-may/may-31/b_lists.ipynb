Question 14 - Write a program to rotate the list to the right by k steps using slicing.<br>
Example - rotate_list([1, 2, 3, 4, 5], 2) -> [4, 5, 1, 2, 3]

```python
input_list_14 = [1, 2, 3, 4, 5]
rotate_step = 2
left_side = input_list_14[:len(input_list_14)-rotate_step:]
print(left_side)
right_side = input_list_14[rotate_step+1::]
print(right_side)
print(right_side + left_side)

#0 1 2 3 4
#2 3 4 0 1
```

Question 15 - Write a program that flattens a list of list into a single list (one level deep)<br>
Example - flatten_list([[1,2], [3,4]]) -> [1, 2, 3, 4]

```python
list_of_list = [
    [1, 2],
    [3, 4]
]
flat_list = [item for inner_list in list_of_list for item in inner_list]
print(flat_list)
```

Question 16 - Write a program that list of unique elements common to two lists.<br>
Example - list_interaction([1, 2, 3], [2, 3, 4]) -> [2, 3]

```python
input_list_1501 = [1, 2, 3]
input_list_1502 = [2, 3, 4]

result = []

set1 = set(input_list_1501)
#set2 = set(input_list_1502)

for num in input_list_1502:
    if num in set1 and num not in result:
        result.append(num)

print(result)
```

Question 17: Write a program to Average of List

```python
input_list_17 = [10, 20, 30, 40, 50, 60]
sum = 0
count = 0
average = 0
for num in input_list_17:
    sum += num
    count += 1
    average = sum/count
print(average)
```

Question 18: Write a program that squares odd numbers (List Comprehension)

```python
input_list_18 = [1, 2 ,3 ,4, 5]

output_list = [num*num for num in input_list_18 if num%2 ==1]
print(output_list)
```

Question 19: Write a program that moves zeroes at the end of the array while maintaining the relative order of the non-zero elements.

```python
input_str19 = [1, 2, 3, 0, 4, 0]

for i in range(len(input_str19)):
    if input_str19[i] == 0:
        input_str19.append(0)
        input_str19.pop(i)
    
print(input_str19)
```
