Question 03 - Write a code that checks for palindrome <br>

```python
input_str03 = "alag"
is_palindrome = True
end_pointer = len(input_str03) -1
start_pointer = 0
while start_pointer < end_pointer:
    if input_str03[start_pointer] != input_str03[end_pointer]:
        print("NOT A PALINDROME")
        is_palindrome = False
        break
    else:
        start_pointer +=1
        end_pointer -=1

if is_palindrome == True:
    print("PALINDROME")
```

Question 5 - How to Insert and delete in dictionary, string, list, tuple and set

```python
# List 
nums = [1, 2, 3]
nums.append(4)      # Insertion at end
print(f"Appending: {nums}")
nums.insert(1, 11)         # At selected
print(f"Adding at the desired position{nums}") 
nums.insert(1,10)
print(f"Adding 2nd number at the desired position{nums}")


nums.pop()          # Deletion at last element
print(f"Number after 1st deletion{nums}")
nums.pop(1)         # Deletion at Index 1
print(f"Number after 2nd deletion{nums}")
nums.remove(2)      # Deletion at Index 2
print(f"Number after 3rd deletion{nums}")
del nums[0]         # Deletion at Index 0
print(f"Number after 4th deletion{nums}")
```

```python
# Dictionary

freq = {}
freq["apple"] = 1
freq["banana"] = 2 
print(f"Inserting into the dictionary{freq}")

freq.pop("apple", None)
print(f"Deletion from the dict{freq}")
```

```python
# Set

s = {1, 2, 3}
print(f"Creation of the set{s}")
s.add(4)
print(f"Updation of the set{s}")

s.remove(2)
print(f"1st Deletion from the Set{s}")
s.discard(3)
print(f"2nd Deletion from the Set{s}")
s.pop()
print(f"3rd Deletion from the Set{s}")
```

Questions 7  - Write a program that remove punctuation

Question 9: Write a program to check for the longest word

Q10. Write a program that checks and returns the first non-repeating character

```python
input_str_10 = "leetcode"
freq = {}
for char in input_str_10:
    freq[char] = freq.get(char, 0) + 1

non_repeat_first_char = ""

for char in freq:
    if freq[char] == 1:
        non_repeat_first_char = char
        break

print(non_repeat_first_char)
```
