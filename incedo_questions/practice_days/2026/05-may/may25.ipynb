Question 01 - Write a function that takes a string as input and returns it reversed. <br>
Example: reverse_string("hello") -> "olleh"

```python
input_str1 = "hello"
def reverse_string(input_str: str) -> str:
    return input_str[::-1]

print(f"The input string is {input_str1} and its reverse is {reverse_string(input_str1)}")
```

Question 02 - Write a function that takes a list of numbers and returns their sum. <br>
Example: sum_list([1, 2, 3, 4]) -> 10

```python
input_list2 = [1, 2, 3, 4]

def sum_list(input_list: list[int]) -> int:
    sum=0
    for num in input_list:
        sum += num
    return sum
    
print(f"The sum of {input_list2} is {sum_list(input_list2)}")
```

Question 03 - Write a function that takes an integer and returns "even" if it's even, "odd" if it's odd. <br>
Example: check_even_odd(5) -> "odd"

```python
input_num3 = 0
def check_even_odd(input_num: int) -> str:
    if input_num == 0:
        return "Nor odd nor even"
    elif input_num%2 ==0:
        return "Even"
    else:
        return "odd"

print(f"{input_num3} is {check_even_odd(input_num3)}")
```

Question 4 : Write a function to compute the factorial of a non-negative intteger.<br>
Example: factorial(5) -> 120


```python
input_num4 = 5
def factorial(input_num: int) -> int:
    fact = 1
    num = 1
    while num <= input_num:
        fact *= num
        num += 1
    return fact
print(f"The factorial of {input_num4} is {factorial(input_num4)}")
```

Question 5: Write a function that counts the number of vowels (a, e, i, o, u) in a string<br>
Example - count_vowels("hello") -> 2

```python
input_str5 = "hello"

def count_vowels(input_str: str) -> int:
    vowel_list = ['a', 'e', 'i', 'o', 'u']
    vowel_count = 0
    for char in input_str:
        if char in vowel_list:
            vowel_count += 1
    return vowel_count

print(f"The vowel count for {input_str5} is {count_vowels(input_str5)}")
```
