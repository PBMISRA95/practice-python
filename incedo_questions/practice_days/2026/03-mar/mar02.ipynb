Question 17 - Write a function that checks whether an integer is prime. <br>
Example: is_prime(11) -> True

```python
input_num17 = 4

def is_prime(input_num: int):
    if input_num < 1:
        return False
    else:
        for num in range(2, int(input_num+1/2)):
            if input_num%num == 0:
                return False
    return True
    

print(is_prime(input_num17))
```

Question 18: Write a function that returns a list of all prime numbers less than or equal to a given limit.<br>
Example: list_primes(10) -> [2, 3, 5, 7]

```python
target_num18 = 11

def list_primes(target: int) -> list[int]:
    list_of_primes = []
    for num in range(2, target+1):
        divisor_count = 0
        for divisor in range(2, num):
            if num%divisor == 0:
                divisor_count += 1
        if divisor_count == 0:
            list_of_primes.append(num)
    return list_of_primes

print(list_primes(target_num18))
```
