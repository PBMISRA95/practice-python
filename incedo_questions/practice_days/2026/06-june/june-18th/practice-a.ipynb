Question 2 - Write a funtion that counts the number of vowels (a, e, i, o, u) in a string.<br>
Example - count_vowels("hello") -> 2

```python
input_str_02 = "hello"
def count_vowels(input_str: str) -> str:
    vowel_count = 0
    vowels = ['a', 'e', 'i', 'o', 'u']
    for char in input_str:
        if char in vowels:
            vowel_count += 1
    return vowel_count

print(f"The vowel count of {input_str_02} is {count_vowels(input_str_02)}")
```

3. Palindrome check <br>
Write a function that checks if a string is a palindrome (case-insenstive).<br>
Example: is_palindrome("Racecar") -> True


```python
input_str_03 = "Racecar"
def is_palindrome(input_str: str) -> str:
    left_pointer = 0
    right_pointer = len(input_str) - 1
    palindrome_status = True
    input_str = input_str.lower()
    while left_pointer < right_pointer:
        if input_str[left_pointer] != input_str[right_pointer]:
            palindrome_status = False
            break
        left_pointer += 1
        right_pointer -= 1
    return palindrome_status

print(f"{input_str_03} is a Palindrome: {is_palindrome(input_str_03)}")
```

```python
input_str_04 = "the cat the dog"
word_freq = {}
def word_frequency(input_str: str) -> dict:
    for word in input_str_04.split(' '):
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
    return word_freq

print(f"The frequency of words in {input_str_04} is \n {word_frequency(input_str_04)}")



#print(input_str_04.split(' '))
```

6. Remove Punctuation <br>
Write a function that removes all punctuation characters from a string. <br>
Example: remove_punctuation("Hi, there!") -> "Hi there"


```python
import string
input_str_06 = "Hi, there!"

def remove_punctuation(input_str: str) -> str:
    new_string = ""
    for char in input_str:
        if char not in string.punctuation:
            new_string += char
    return new_string

print(f"Input String: {input_str_06} is\nNo punctuation: {remove_punctuation(input_str_06)}")
```

Example: are_anagrams("Listen","Silent") -> True


```python
input_str_71 = "Listen"
input_str_72 = "Silent"
def are_anagrams(input_str1: str, input_str2: str) -> bool:
    is_anagram = True
    input_str1 = input_str1.lower()
    input_str1 = sorted(input_str1)
    input_str2 = input_str2.lower()
    input_str2 = sorted(input_str2)
    if len(input_str1) == len(input_str2):
        for i in range(0, len(input_str1)):
            if input_str1[i] != input_str2[i]:
                is_anagram = False
                break
            else:
                pass
                    
    else:
        is_anagram = False
    
    return is_anagram

print(are_anagrams(input_str_71,input_str_72))
```
