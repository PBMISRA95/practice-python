1. Sum of List<br>
Write a function that takes a list of numbers and returns their sum.<br>
Example: sum_list([1, 2, 3, 4] -> 10)

```python
input_list_01 = [1, 2 ,3 ,4]
def sum_list(input_list:list[int]) -> int:
    sum = 0
    for num in input_list:
        sum += num
    return sum

print(f"The sum of {input_list_01} is {sum_list(input_list_01)}")
```

4. Rotate List<br>
Write a function that rotates a list to the right by k positions using slicing.<br>
Example: rotate_list([1, 2, 3, 4, 5], 2) -> [4, 5, 1, 2, 3]

```python
#[k+1:len(list):1] + [:k:]
input_list_4 = [1, 2, 3, 4, 5]
k = 2

def rotate_list(input_list:list[int], k) -> list[int]:
    return input_list[k+1:len(input_list):] + input_list[:k+1:]

print(f"{input_list_4} is {rotate_list(input_list_4,2)}")
```

5. Flatten Nested List<br>
Write a function that flattens a list of lists into a single list. <br>
Example: flatten_list([[1,2],[3,4]]) -> [1, 2, 3, 4]

```python
list_of_lists = [
    [1,2],
    [3,4]
]

def flatten_list(nested_list):
    flat_list = []

    for sublist in nested_list:
        for item in sublist:
            flat_list.append(item)
    
    return flat_list

print(flatten_list(list_of_lists))
```

9. Move Zeroes <br>
Write a function that moves all zeroes to the end of the list while maintaining the order of non-zero elements.<br>
Example: move_zeroes([0, 1, 0, 3, 12]) -> [1, 3, 12, 0, 0]

```python
input_list_9 = [0, 1, 0, 3, 12]
def move_zeroes(input_list: list[int]) -> list[int]:
    zeroes_list = []
    output_list = []
    for num in input_list:
        if num == 0:
            zeroes_list.append(0)
        else:
            output_list.append(num)
    return output_list + zeroes_list
        
print(f"{input_list_9} is {move_zeroes(input_list_9)}")     
```
