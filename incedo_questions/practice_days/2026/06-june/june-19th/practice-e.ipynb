1. Prime Check<br>
Write a function that checks whether an integer is prime.<br>
Example: is_prime(11) -> True

```python
input_num = 11
def is_prime(input_num: int)-> bool:
    for i in range(2, int(input_num/2) + 1):
        if input_num % i == 0:
            return False
    return True    

print(is_prime(input_num))
```

Example: list_primes(10) -> [2,3,5,7]

```python
input_num_2 = 10
def list_primes(input_num:int)-> list[int]:
    list_of_primes = []
    is_prime = True
    for num in range(2, input_num+1):
        is_prime = True
        for i in range(2,num):
            if num % i == 0:
                is_prime = False
                break

        if is_prime == True:
            list_of_primes.append(num)
    return list_of_primes



print(list_primes(input_num_2))
```

3. GCD<br>
Write a function that returns the greatest common divisor of two integers.<br>
Example: gcd(48, 18) -> 6

```python
def gcd(input_num1: int, input_num2: int) -> int:
    if input_num1 > input_num2:
        target = input_num2 + 1
    else:
        target = input_num1 + 1
    max_divisor = 0
    for i in range(1, target):
        if input_num1 % i == 0 and input_num2 % i == 0:
            if i > max_divisor:
                max_divisor = i
    
    return max_divisor


print(gcd(48,24))
```

5. Fibonacci Number<br>
Write a function that returns the nth Fibonacci number(0-indexed).<br>
Example: fibonacci(7) -> 13

```python
def fibonacci(num: int) -> int:
    if num ==0:
        return 0
    if num == 1:
        return 1

    a, b = 0, 1
    for _ in range(2, num + 1):
        a, b = b, a + b
    
    return b

print(fibonacci(7))
```

7. Reverse Number <br>
Write a function that reverses the digits of an integer.<br>
Example: reverse_number(1234) -> 4321

```python
def reverse_number(num: int) -> int:
    remainder = 0
    new_number = 0
    quotient = num

    while quotient > 0:
        remainder = quotient%10
        quotient = int(quotient / 10)
        new_number = new_number * 10 + remainder
    
    return new_number

print(str(reverse_number(1234)))

```
