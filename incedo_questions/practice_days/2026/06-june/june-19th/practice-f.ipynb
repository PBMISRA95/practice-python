1. Two Sum<br>
Write a function that returns the indices of two number whose sum equals a target value. <br>
Example: two_sum([2, 7, 11, 15], 9) -> [0, 1]

```python
input_list1 = [2, 15, 11, 15, 7]
target = 9

def two_sum(input_list: list[int], target) -> list[int]:
    left_pointer = 0
    n = len(input_list)
    right_pointer = n-1
    for i in range(n):
        current_sum = 0
        current_sum = input_list[left_pointer] + input_list[right_pointer]
        if current_sum == target:
            return [left_pointer, right_pointer]
        elif current_sum > target:
            right_pointer -= 1
        elif current_sum < target:
            left_pointer += 1
    
print(two_sum(input_list1, target))
```
