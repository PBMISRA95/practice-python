1. Even/Odd <br>
Write a function that returns "even" if a number is even and "odd" if it is odd. <br>
Example: check_even_odd(5) -> "odd"

```python
input_num = 0
def check_even_odd(input_num: int) -> str:
    if input_num != 0:
        if input_num % 2 == 0:
            return "even"
        else:
            return "odd"
    else:
        return "Zero"

print(check_even_odd(input_num))
```

2. Factorial<br>
Write a function that computes the factorial of a non-negative integer.<br>
Example: factorial(5) -> 120

```python
input_num_2 = 5
def factorial(input_num: int)-> int:
    fact = 1
    for i in range(1,input_num+1):
        fact *= i
    return fact

print(factorial(input_num_2))
```

3. Second Largest Element <br>
Write a function that returns the second-largest unique element in a list.<br>
Example: second_largest([4, 1, 7, 7, 3]) -> 4



```python
input_list = [4,1,7,7,3]
def second_largest(input_list:list[int]) -> int:
    output_list = sorted(list(set(input_list)))
    return output_list[len(output_list)-2]

print(second_largest(input_list))
```

4. Leap Year<br>
Write a function that determines whether a year is a leap year.<br>
Example: is_leap_year(2000) -> True

```python
# 1992 1996 2000 - 2004 2008 2012
input_year = 2012
def is_leap_year(input_year: int)-> bool:
    if input_year % 100 == 0:
        if input_year % 400 == 0:
            return True
        elif input_year % 400 !=0:
            return False
    elif input_year % 100 !=0:
        if input_year % 4 == 0:
            return True
        else: 
            return False
 
print(is_leap_year(input_year))
```
