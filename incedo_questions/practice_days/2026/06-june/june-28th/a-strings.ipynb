1. Strings<br>
Write a function that takes a string as input and returns it reversed.<br>
Example: reverse_string("hello") -> "olleh"

```python
input_str = "hello"
def rev(input_str: str) -> str:
    return input_str[::-1]

print(rev(input_str))
```

2. Count Vowels<br>
Write a function that counts the number of vowels (a, e, i, o, u) in a string.<br>
Example - count_vowels("hello") -> 2

```python
input_str2 = "helloo"
def count_vowels(input_str: str) -> int:
    vowel_count = 0
    for char in input_str:
        if char in ['a', 'e', 'i', 'o', 'u']:
            vowel_count += 1
        
    return vowel_count

print(count_vowels(input_str2))
```

3. Palindrome Check <br>
Write a function that checks if a string is palindrome (case-insensitive). <br>
Example: is_palindrome("Racecar") -> True

```python
input_str3 = "RacecaR"

def is_palindrome(input_str: str) -> bool:
    palindrome_status = True
    left_pointer = 0
    right_pointer = len(input_str) - 1
    input_str = input_str.lower()
    for i in range((len(input_str)// 2)):
        if input_str[left_pointer] != input_str[right_pointer]:
            palindrome_status = False
        left_pointer += 1
        right_pointer -= 1
    
    return palindrome_status

print(is_palindrome(input_str3))
```

4. Word Frequency <br>
Write a function that takes a string and returns a dictionary containing the frequency of each word.<br>
Example: word_frequency("the cat the dog") -> {"the": 2, "cat": 1, "dog": 1}

```python
input_str4 = "the cat the god"
def word_frequency(input_str: str) -> str:
    word_list = input_str.split()
    freq_dict = {}
    for word in word_list:
        freq_dict[word] = freq_dict.get(word, 0) + 1
    return freq_dict

print(word_frequency(input_str4))
```

5. Character Frequency<br>
Write a frequency that returns a dictionary containing the frequency of each character in a string.<br>
Example: count_chars("hello")

```python
input_str5 = "hello"

def count_chars(input_str: str) -> int:
    freq = {}
    for char in input_str:
        freq[char] = freq.get(char, 0) + 1
    return freq

print(count_chars(input_str5))
```

6. Join Words<br>
Write a function that takes a list of words and joins them into a single string using spaces.<br>
Example: join_words(["hello","world"]) -> "hello world"

```python
list_of_words = ["hello", "world"]

def join_words(word_list: list[str] ) -> str:
    joined_str = ""
    for word in word_list:
        joined_str += " " + word

    return joined_str

print(join_words(list_of_words))
```

7. Remove Punctuation<br>
Write a function that removes all punctuation characters from a string.<br>
Example: remove_punctuation("Hi, there!") -> "Hi there"

```python
import string

input_str7 = "hello world !!"

def remove_punctuation(input_str: str) -> str:
    cleaned_word =""
    punctuation = str(string.punctuation)
    for char in input_str:
        if char not in punctuation:
            cleaned_word += char
    
    return cleaned_word

print(remove_punctuation(input_str7))
```

8.  Anagram Check<br>
Write a function that checks whether two strings are anagrams of each other (case-insensitive, ignoring spaces). <br>
Example: are_anagrams("Listen", "Silent") -> True

```python
def are_anagrams(input_str1: str, input_str2: str) -> bool:
    anagram_status = True
    if len(input_str1) != len(input_str2):
        anagram_status = False
    else:
        for i in range(len(input_str1)):
            if sorted(input_str1.lower()) != sorted(input_str2.lower()):
                anagram_status = False
    return anagram_status

print(are_anagrams("Listepn","Silpent"))
```

9. Longest Word<br>
Write a function that returns the longest word in a sentence.<br>
Example: longest_word("I love machine learning") -> "learning"

```python
def longest_word(input_str: str)-> str:
    word_length = {}
    for word in input_str.split():
        word_length[word] = len(word)
    
    return max(word_length.keys())

print(longest_word("I love machine learningpo"))
```

10. First Non-Repeating Character<br>
Write a function that returns the first character that appears only once in a string.<br>
Example: first_non_repeating("swiss") -> "w"

```python
def first_non_repeating(input_str: str) -> str:
    freq = {}
    for ch in input_str:
        freq[ch] = freq.get(ch, 0) + 1
    for key,value in freq.items():
        if value == 1:
            return key

print(first_non_repeating("swiss"))
```
