1. Sum of List<br>
Write a function that takes a list of numbers and returns their sum.<br>
Example: sum_list([1,2,3,4] -> 10)

```python
input_list = [1, 2, 3, 4]
def sum_list(input_list):
    sum = 0
    for num in input_list:
        sum += num
    return sum

print(sum_list(input_list))
```

2. Find Maximum<br>

Example: find_max([3, 1, 4, 1, 5]) -> 5

```python
input_list_2 = [3, 1, 4, 1, 5]
def find_max(input_list):
    return max(input_list)

print(find_max(input_list_2))
```

3. Remove Duplicates<br>
Write a function that removes duplicate elements from a list.<br>
Example: remove_duplicates([1, 2, 2, 3, 4, 4]) -> [1, 2, 3, 4]

4. Rotate List <br>
Write a function that rotates a list to the right by k positions using slicing. <br>
Example: rotate_list([1, 2, 3, 4, 5], 2) -> [4, 5, 1, 2, 3]

```python
input_list4 = [1, 2, 3, 4, 5]
k = 2

def rotate_list(input_list, k):
    return input_list[k+1:len(input_list):] + input_list[:k+1:]

print(rotate_list(input_list4, k))
```

5. Flatten Nested List<br>
Write a function that flattens a list of lists into a single list.<br>
Example: flatten_list([[1, 2], [3, 4]]) -> [1, 2, 3, 4]

```python
input_list5  = [
    [1, 2],
    [3, 4]
    ]
def flatten_list(list1:list[list]) -> list:
    output_list = []
    for list in list1:
        for num in list:
            output_list.append(num)
    
    return output_list

print(flatten_list(input_list5))
```

6. List Intersection<br>
Write a function that returns unique elements common to two lists.<br>
Example: list_intersection([1, 2, 3], [2, 3, 4]) -> [2, 3]

```python
def list_intersection(input_list1: list[int], input_list2: list[int]) -> list[int]:
    output_list = []
    for num in input_list1:
        if num in input_list2:
            output_list.append(num)
    
    return output_list

print(list_intersection([1, 2, 3], [2, 3, 4]))
```

7. Average of list<br>
Write a function that returns the arithmetic mean of a list. Return None for an empty list.<br>
Example: average_list([4, 8]) -> 6.0

```python
input_list1 = [4,8,12]
def average_list(input_list: list[int]) -> int:
    sum = 0
    steps = 1
    average = 0
    for num in input_list:
        sum += num
        average = sum/steps
        steps += 1
    
    return average

print(average_list(input_list1))
```

8. Square Odd Numbers (List Comprehension)<br>
Write a function that squares odd numbers in a list using list comprehension.<br>
Example: square_odds([1, 2, 3, 4]) -> [1, 2, 9, 4]

```python
input_list8 = [1, 2, 3, 4]

def square_odds(input_list: list[int]) -> list[int]:
    output_list = [num * num if num % 2 == 1 else num for num in input_list]
    return output_list

print(square_odds(input_list8))
```

9. Move Zeroes<br>
Write a function that moves all zeroes to the end of the list while maintaining the order of non-zero elements.<br>
Example: move_zeroes([0, 1, 0, 3, 12]) -> [1, 3, 12, 0, 0]

```python
input_list9 = [0, 1, 0, 3, 12]
def move_zeroes(input_list: list[int]) -> list[int]:
    zero_list = []
    non_zero_list = []
    output_list = []
    for num in input_list:
        if num == 0:
            zero_list.append(num)
        else:
            non_zero_list.append(num)

    output_list = zero_list + non_zero_list

    return output_list

print(move_zeroes(input_list9))
```
