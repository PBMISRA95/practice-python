1. Merge Two Dictionaries<br>
Write a function that merges two dictionaries. If duplicate keys exist, values from the second dictionary should override the first<br>
Example - merge_dicts({'a': 1}, {'a': 2, 'b': 3}) -> {'a': 2, 'b': 3}

```python
input_dict1 = {'a': 1} 
input_dict2 = {'a': 2, 'b': 3}
def merge_dicts(dict1: dict, dict2: dict) -> dict:
    output_dict = {}
    for key, value in dict1.items():
        output_dict[key] = value
    
    for key, value in dict2.items():
        if key in output_dict:
            output_dict[key] = value
        else:
            output_dict[key] = value
    
    return output_dict

merge_dicts(input_dict1, input_dict2)

"""

def merge_dicts(dict1: dict, dict2: dict) -> dict:
    result = dict1.copy()
    result.update(dict2)
    return result
"""
```

2. Frequency Count of Elements<br>
Write a function that returns the frequency count of elements in a list.<br>
Example: frequency_count([1, 2, 2, 3]) -> {1: 1, 2: 2, 3: 1}

```python
input_list2 = [1, 2, 2, 3]
def frequency_count(input_list: list[int]) -> dict:
    freq = {}
    for num in input_list:
        freq[num] = freq.get(num, 0) + 1
    return freq

print(frequency_count(input_list2))
```

3. Sort list of Dictionaries by Key<br>
Write a function that sorts a list of dictionaries by the "name" key in ascending order.<br>
Example: [<br>
    {"name" : "Bob"},<br>
    {"name" : "Alice"}<br>
]   -><br>
 [<br>
    {"name" : "Alice"},<br>
    {"name" : "Bob"}<br>
 ]<br>

```python
list_of_dict = [
    {"name": "Bob"},
    {"name": "Alice"}
]

def sort_by_name(people: list[dict]) -> list[dict]:
    return sorted(people, key=lambda x: x["name"])

print(sort_by_name(list_of_dict))
    
```
