Question 1 - Write a function that checks whether an integer is prime<br>
Example: is_prime(11) -> True

```python
input_num_1 = 11
# logic 
# is_prime = true (if number of divisors more than 2)
# num_of_divisors = 0 

def is_prime(num:int) -> bool:
    is_prime_number = True
    num_of_divisors = 1
    for i in range(2, int(num/2)+1):
        if num%i == 0:
            num_of_divisors +=1
        if num_of_divisors > 1:
            is_prime_number = False
            break
        i += 1
    return is_prime_number
    
print(f"{input_num_1} is {is_prime(input_num_1)}")
```

Question 2 - Write a function that returns all prime numbers less than or equal to a given limit<br>
Example: is_primes(10) -> [2, 3, 5, 7]

```python
# LOGIC 
# prime number logic under this number 
# 1, 2, 3, 4, 5, 7, 8, 9, 10
# list_of_primes = []
# is_prime 
# number_of_divisors 
# 

number_limit = 12
list_of_primes = []
divisor_number = 0
def is_primes(num_limit: int) -> list[int]:
    for number in range(2,num_limit+1):
        divisor_number = 0
        for num in range(2,int(number/2)+1):
            if number%num == 0:
                divisor_number += 1
                if divisor_number > 1:
                    break
        if divisor_number == 0:
            list_of_primes.append(number)
    return list_of_primes

print(f"{number_limit} has the following prime number{is_primes(number_limit)}")        
            
```

Question 3 - Write a function that returns the greatest common divisor of two integers<br>
Example: gcd(48, 18) -> 6

```python
# LOGIC
# max_divisor = half of smallest number
# From max_divisor until 2 find the 1st divisor and then return it

input_num_2 = 48
input_num_3 = 18
max_num_to_check = 0
#max_divisor = 1

def gcd(num1: int, num2: int) -> int:
    max_divisor =1
    if num1 > num2:
        max_num_to_check = num2
    else:
        max_num_to_check = num1
    for i in range(int(max_num_to_check/2)+1, 2, -1):
        if num1 % i == 0 and num2 % i== 0:
            max_divisor = i
            break
        else:
            i-=1
    return max_divisor


print(f"{input_num_2} and {input_num_3} have the following GCD: {gcd(input_num_2, input_num_3)}")
```
