1. Two Sum<br>
Write a function that returns the indices of two numbers whose sum equals a target value.<br>
Example: two_sum([2, 7, 11, 15], 9) -> [0, 1]

```python
input_list1 = [2, 7, 11, 15]
target = 9
def two_sum(list1: list, target: int) -> list[int]:
    left_pointer = 0
    right_pointer = len(list1) - 1
    for i in range(len(list1)):
        if target == list1[right_pointer] + list1[left_pointer]:
            return [left_pointer, right_pointer]
        elif target > list1[right_pointer] + list1[left_pointer]:
            left_pointer += 1
        elif target < list1[right_pointer] + list1[left_pointer]:
            right_pointer -= 1
    
print(two_sum(input_list1, target))    
```

2. Valid Parentheses<br>
Write a function that checks whether a string containing brackets is valid. <br>
Example: is_valid_parentheses("()[]{}") -> True

```python
def is_valid_parentheses(list1: list) -> bool|str:
    stack = []
    opening_brackets = ['{', '(', '[']
    for char in list1:
        if char in opening_brackets:
            stack.append(char)
        else:
            if len(stack) == 0:
                return "Stack is Null"
            last_element = stack.pop()
            if char == ')' and last_element != '(':
                return False
            if char == '}' and last_element != '{':
                return False
            if char == ']' and last_element != '[':
                return False
        
    if len(stack) == 0:
        return True

print(is_valid_parentheses("()[]}"))
```

```python
# Code 2

def is_valid_parentheses(list1: list) -> bool|str:
    stack = []
    correct_brackets = {
        '}': '{',
        ')': '(',
        ']': '['
        }
    for char in list1:
        if char in correct_brackets.values():
            stack.append(char)
        elif char in correct_brackets.keys():
            if not stack or stack.pop() != correct_brackets[char]:
                return False
          

    return len(stack) == 0
        
print(is_valid_parentheses("()[]{}"))
```
