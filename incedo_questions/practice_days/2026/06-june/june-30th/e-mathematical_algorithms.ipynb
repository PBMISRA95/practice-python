1. Prime Check <br>
Write a function that checks whether an integer is prime.<br>
Example: is_prime(11) -> True

```python
input_num = 6
def is_prime(num:int) -> bool:
    for i in range(2, num//2 + 1):
        if num%i == 0:
            return False
    return True

print(is_prime(input_num))
```

2. List of Primes<br>
Write a function that returns all prime numbers less than or equal to a given limit.<br>
Example: list_primes(10) -> [2, 3, 5, 7]

```python
input_num = 20
def list_primes(num: int) -> list[int]:
    prime_list = []
    for i in range(2,num+1):
        is_prime = True 
        for div in range(2,i//2 + 1):
            if i% div == 0:
                is_prime = False
                break
        if is_prime == True:
            prime_list.append(i)
    return prime_list
        

print(list_primes(input_num))
                
```

3. GCD<br>
Write a function that returns the greatest common divisor of two integers. <br>
Example: gcd(48,18) -> 6

```python
def gcd(num1, num2):
    if num1 > num2:
        smallest = num2
        if num1 % smallest == 0:
            print("first if")
            return num2
    else:
        smallest = num1
        if num2 % smallest:
            print("second if")
            return num1

    for i in range(smallest//2+1, 0, -1):
        if num1%i == 0 and num2%i == 0:
            print("inside for loop")
            return i
    print("last line of function")

print(gcd(48,480))
```

4. LCM<br>
Write a function that returns the least common multiple of two integers.<br>
Example: lcm(4, 6) -> 12


```python
def lcm(num1: int, num2: int) -> int:
    max_multiple = num1 * num2
    if num1 > num2:
        biggest = num1 
        if num1 % num2 == 0:
            return biggest
        
    elif num2 > num1:
        biggest = num2
        if num2 % num1 == 0:
            return biggest
        
    for i in range(biggest,max_multiple,1):
        if i%num1 ==0 and i%num2 ==0:
            return i
        
    return max_multiple

print(lcm(4,8))

```

5. Fibonacci Number <br>
Write a function that returns the nth Fibonacci number (0-indexed).<br>
Example: fibonacci(7) -> 13

```python
def fibonacci(num: int) -> int:
    f =[0,1]
    sum = 0
    for i in range(2, num+1):
        prev = i-1
        prev2 = i-2
        sum = f[prev] + f[prev2]
        f.append(sum)
    return f[num]

print(fibonacci(7))

# Fibonacci index  = 0 1 2 3 4 5 6 7
# Fibonacci Series = 0 1 1 2 3 5 8 13
```

6. Generate Fibonacci Series <br>
Write a function that generates the first n Fibonacci numbers.<br>
Example: generate_fibonacci(5) -> [0, 1, 1, 2, 3]

```python
def generate_fibonacci(num: int) -> list[int]:
    fibonacci = [0, 1]
    for i in range(2,num + 1):
        new_element = fibonacci[i-2] + fibonacci[i-1]
        fibonacci.append(new_element)
    return fibonacci

print(generate_fibonacci(5))
```

7. Reverse a Number <br>
Write a function that reverses the digits of an integer.<br>
Example: reverse_number(1234) -> 4321

```python
def reverse_number(num:int) -> int:
    reversed_num = 0
    remainder = 0
    while num > 0:
        remainder = num % 10
        reversed_num = reversed_num * 10 + remainder
        num = num // 10
    return reversed_num

print(reverse_number(1234))
```
