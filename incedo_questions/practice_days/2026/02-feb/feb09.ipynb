Question 10 - Write a function that takes a list of words and joins them into a single string with spaces using join() <br>
Example: join_words(["hello","world"]) -> "hello world"

```python
input_str10 = ["hello", "world"]

def join_words(input_str: list):
    return " ".join(input_str)

print(join_words(input_str10))
```
