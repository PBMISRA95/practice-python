Question 2 - Write a function that takes a list of numbers and returns their sum.<br>
Example: sum_list([1, 2, 3, 4]) -> 10

```python
input_num = [1,2,3,4]

def sum_list(input_num: list[int] | None = None) -> int:
    if input_num is None:
        input_num = []
    total = 0
    for num in input_num:
        total += num
    return total

print(sum_list(input_num))
```

Question 3: Write a function that takes an integer and returns "even" if it's even, "odd" if it's odd.<br>
Example - check_even_odd(5) -> "odd"



```python
input_num03 = 0
def check_even_odd(input_num: int) -> str:
    if input_num == 0:
        return "ZERO NOT EVEN OR ODD"
    else:
        if (input_num % 2) == 0:
            return "EVEN"
        else:
            return "ODD"

print(check_even_odd(input_num03))
```

Question 4: Write a function to compute the factorial of a non-negative integer. 
Example: factorial(5) -> 120

```python
input_num04 = 6
def factorial(input_num: int) -> int:
    fact = 1
    if input_num > 0:
        while input_num !=0:
            fact *= input_num
            input_num -= 1
    return fact

print(factorial(input_num04))
```

Question 5: Write a function that counts the number of vowels (a, e, i, o, u) in a string.<br>
Example: count_vowels("hello") -> 2


```python
input_str05 = "hello"

def count_vowels(input_str: str):
    vowel_count = 0
    vowel_list = ['a', 'e', 'i', 'o', 'u']
    for char in input_str:
        if char in vowel_list:
            vowel_count += 1
    return vowel_count

print(count_vowels(input_str05))
        
```

Question 6: Write a function that takes a list of numbers and returns the maximum value. <br>
Example: find_max([3, 1, 4, 1, 5]) -> 5

```python
input_num06 = [3, 1, 4, 1, 10]

def find_max(input_num):
    # max = max(input_num)
    max = input_num[0]
    for num in input_num:
        if max < num:
            max = num
    return max

print(find_max(input_num06))

```

Question 7 - Write a functio nthat checks if a string is a palindrome (case-insenstive)<br>
Example - is_palindrome("Racecar") -> True

```python
input_str07 = "RacEcar"
def is_palindrome(input_str):
    input_str = input_str.lower()
    while(input_str):
        if input_str[::] != input_str[::-1]:
            return False
        else:
            return True

print(is_palindrome(input_str07))
```
