Question 14 - Write a function that takes a list of dictionaries and sorts it by the "name" key in ascending order. <br>
Example: data = [{"name": "Bob", "age": 25}, {"name": "Alice", "age": 30}] -> [{"name": "Alice"}, {"name": "Bob", "age" : 25}]


```python
list_of_dictionaries = [
    {"name": "Bob", "age" : 25},
    {"name": "Alice", "age" : 30}
]

def sort_by_name(input_list: list[dict]) -> list[dict]:
    return sorted(input_list, key=lambda d: d["name"])

print(sort_by_name(list_of_dictionaries))
```
