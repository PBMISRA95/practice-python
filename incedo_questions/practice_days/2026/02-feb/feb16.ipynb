Question 14 - Write a function that takes a list of dictionaries and sorts it by the "name" key in ascending order. <br>
Example: data = [{"name": "Bob", "age": 25}, {"name": "Alice", "age": 30}] -> [{"name": "Alice"}, {"name": "Bob", "age" : 25}]


```python
list_of_dictionaries = [
    {"name": "Bob", "age" : 25},
    {"name": "Alice", "age" : 30}
]

def sort_by_name(input_list: list[dict]) -> list[dict]:
    return sorted(input_list, key=lambda d: d["name"])

print(sort_by_name(list_of_dictionaries))
```

Question 15: Write a function that takes a list of numbers and return their arithematic mean (return None for an empty list). <br>
average_list([4,8]) -> 6.0

```python
input_list15 = [4,8,10,12,15]
def average_list(input_list: list[int]) -> int:
    if input_list:
        average = 0
        sum = 0
        total_numbers = 0
        for num in input_list:
            sum += num
            total_numbers += 1
            average = float(sum/total_numbers)
        return average
    else:
        return None

print(average_list(input_list15))
```

Question 16: Write a function that returns the second-largest unique number in a list (raise an error if the list has fewer than two unique values). <br>
Example: second_largest([4, 1, 7, 7, 3]) -> 4

```python
input_list16 = [4, 1, 7, 7, 3]
#input_list16 = [7,7]
def second_largest(input_list: list[int]) -> int:
    if len(input_list) < 3:
        #raise Exception("Error: less than 2 element")
        if len(input_list) != len(set(input_list)):
            raise Exception("Error: Less than 2 unique values")
    else:
        max_element = max(input_list)
        second_largest = 0
        for num in input_list:
            if num == max_element:
                continue
            elif second_largest < num:
                second_largest = num
        return second_largest
    
print(second_largest(input_list16))


```
