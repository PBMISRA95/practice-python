Question 11 - Write a function that returns a dictionary of character counts in a string. <br>
Example: count_chars("hello") -> {'h': 1, 'e': 1, 'l': 2, 'o': 2}

```python
input_str11 = "hello"
def count_chars(input_str: str) -> dict:
    char_count = {}
    for char in input_str:
        if char in char_count:
            char_count[char] += 1
        elif char not in char_count:
            char_count[char] = 1
    return char_count


print(count_chars(input_str11))
```

Question 12: Write a function that square numbers in a list if they are odd, using a list comprehension with a conditional.<br>
Example: square_odds([1,2,3,4]) -> [1, 2, 9, 4]

```python
input_list = [1,2,3,4]
def square_odds(input_list: list[int]) -> list[int]:
    output_list = [num * num if num%2 ==1 else num for num in input_list]
    return output_list

print(square_odds(input_list))
```

Question 13: Write a function that rotates a list to the right by k steps using slicing.<br>
Example: rotate_list([1, 2, 3, 4, 5], 2) -> [4, 5, 1, 2 ,3]

```python
input_list13 = [1, 2, 3, 4, 5]
rotating_steps = 2
def rotate_list(input_list: list[int], steps:int) -> list[int]:
    n = len(input_list)
    return input_list[n - steps:n-steps:-1]

print(rotate_list(input_list13, rotating_steps))

```
