Question 11 - Write a function that returns a dictionary of character counts in a string. <br>
Example: count_chars("hello") -> {'h': 1, 'e': 1, 'l': 2, 'o': 2}

```python
input_str11 = "hello"
def count_chars(input_str: str) -> dict:
    char_count = {}
    for char in input_str:
        if char in char_count:
            char_count[char] += 1
        elif char not in char_count:
            char_count[char] = 1
    return char_count


print(count_chars(input_str11))
```
