Question 8 - Write a function that removes duplicates from a list and returns the result.<br>
Example: remove_duplicates([1, 2, 2, 3, 4, 4]) -> [1, 2, 3, 4]


```python
input_list08 = [1, 2, 2, 3, 4, 4]

def remove_duplicates(input_list):
    output_list = []
    for n in input_list:
        if n not in output_list:
            output_list.append(n)
    return output_list

print(remove_duplicates(input_list08))
```
