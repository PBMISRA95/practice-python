Question 8 - Write a function that removes duplicates from a list and returns the result.<br>
Example: remove_duplicates([1, 2, 2, 3, 4, 4]) -> [1, 2, 3, 4]


```python
input_list08 = [1, 2, 2, 3, 4, 4]

def remove_duplicates(input_list):
    output_list = []
    for n in input_list:
        if n not in output_list:
            output_list.append(n)
    return output_list

print(remove_duplicates(input_list08))
```

Question 9 - Write a function that takes a string and returns a dictionary with the frequency of each word.<br>
Example: word_frequency("the cat the dog") -> {"the":2, "cat":1, "dog": 1}


```python
input_str09 = "the cat the god the"
def word_frequency(input_str):
    word_freq = {}
    word = ""
    for char in input_str:
        if char != " ":
            word = word + char
        elif char == " ":
            if word not in word_freq:
                word_freq[word] = 1
            elif word in word_freq:
                word_freq[word] +=1
            word = ""
    if word:
        if word not in word_freq:
            word_freq[word] = 1
        elif word in word_freq:
            word_freq[word] +=1
    return word_freq
    
print(word_frequency(input_str09))
```
