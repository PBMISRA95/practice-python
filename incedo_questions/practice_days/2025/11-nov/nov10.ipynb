```python
# Q13. Write a function that rotates a list to the right by k steps using slicing.
# Example - rotate_list([1,2,3,4,5], 2) -> [4,5,1,2,3]
def rotate_list(input_list: list[int], rotate_steps: int) -> list[int] | None:
    break_index = len(input_list)-rotate_steps
    output_list = input_list[break_index::] + input_list[:break_index:]
    return output_list

print(rotate_list([1,2,3,4,5],2))
```
