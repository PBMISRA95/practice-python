```python
# Q13. Write a function that rotates a list to the right by k steps using slicing.
def rotate_list(input_steps: int, input_list : list[int]) -> list[int]:
    output_list = input_list[-input_steps::] + input_list[:-input_steps:]
    return output_list

print(rotate_list(2, [1, 2, 3, 4, 5]))
```
